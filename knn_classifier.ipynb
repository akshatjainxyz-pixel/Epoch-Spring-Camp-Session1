```python
import numpy as np
```

```python
#Collecting data
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]
```

```python
#Making new column for label encoding( pre processing data)
for i in data:
    if i[3]=='Apple':
        i.append(0)
    elif i[3]=='Banana':
        i.append(1)
    else:
        i.append(2)
```

```python
#Separating data into X and y
data = np.array(data)
X = np.array(data[:, :3], dtype= float)
y = np.array(data[:, 4], dtype= int)   
```

```python
#Defining Euclidean Distance Function
def Distance(x,y):
    n=len(x)
    sum=0
    for i in range(n):
        sum= sum+ np.square(x[i]- y[i])
    return np.sqrt(sum)
    
```

```python
class KNN:
    def __init__(self, k=3):
        self.k= k
        self.X_train= None
        self.y_train= None
    def fit(self, X, y): #To store training data
        self.X_train = X
        self.y_train = y
    def predict_one(self, x):
        dist=[]   # dist will store distance of x from each training datapoint AND the index of that training datapoint
        for i in range(len(self.X_train)):
            dist.append([Distance(self.X_train[i], x),i])  #Storing distance, index in the list dist
            dist.sort(key=lambda d: d[0], reverse=False)   #Sorting distance in ascending order
        apple=0 #If an apple is one of the K nearest neighbours, value of apple increases by 1. Similarly for other variables
        banana=0
        orange=0
        for i in range(self.k):
            a= dist[i][1]
            b= self.y_train[a]
            if b==0:
                apple= apple+1
            elif b==1:
                banana = banana+1
            else:
                orange= orange+1
        if apple>=banana and apple>=orange:
            return 'Apple'
        elif banana>=apple and banana>=orange:
            return 'Banana'
        else:
            return 'Orange'
    def predict(self, X_test): #Predicts result for several testing points
        X_result=[]
        for i in X_test:
            X_result.append(self.predict_one(i))
        return X_result
```

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])
model1= KNN(3) #Taking k as 3
model1.fit(X, y)
ans1= model1.predict(test_data)
print(ans1)

model2=KNN(1)  #Taking k as 1
model2.fit(X, y)
ans2= model2.predict(test_data)
print(ans2)

model3=KNN(5) #Taking k as 5
model3.fit(X, y)
ans3= model3.predict(test_data)
print(ans3)


#When we took k =1 and 3, we got the expected output. But when we took k as 5,  the majority closest neighbours for the 3rd test point were apples
#and this gave us wrong results. Thus we can say that if we take a very large value of k, we might be underfitting the data.
```
